A continuació s'eliminen aquelles columnes que no són rellevants per l'estudi, conservant únicament les variables d'interès. 

```python
#Creació d'una nova columna amb la data ['Date'] per poder eliminar les columnes 'Year', 'Month' i 'DayofMonth'

from datetime import datetime
flights['Date'] = flights.apply(lambda row: datetime(row['Year'], row['Month'], row['DayofMonth']), axis=1)
```

```python
#Eliminació de les columnes que no són d'interès

flights = flights.drop("Unnamed: 0", axis=1) #No aporta informació.
flights = flights.drop("Year", axis=1) #Informació ja inclosa a la columna 'Date'
flights = flights.drop("Month", axis=1) #Informació ja inclosa a la columna 'Date'
flights = flights.drop("DayofMonth", axis=1) #Informació ja inclosa a la columna 'Date'
flights = flights.drop("DepTime", axis=1) #La columna d'interès és la que fa referència a l'hora prevista de sortida. La informació d'aquesta columna ja es pot extreure de la columna DepDelay.
flights = flights.drop("ArrTime", axis=1) #La columna d'interès és la que fa referència a l'hora prevista d'arribada. La informació d'aquesta columna ja es pot extreure de la columna ArrDelay.
flights = flights.drop("ActualElapsedTime", axis=1) #No aporta informació rellevant per l'anàlisis.
flights = flights.drop("CRSElapsedTime", axis=1) #No aporta informació rellevant per l'anàlisis.
flights = flights.drop("TaxiIn", axis=1) #No aporta informació rellevant per l'anàlisis, ja que no permet explicar el retard.
flights = flights.drop("TaxiOut", axis=1) #No aporta informació rellevant per l'anàlisis, ja que no permet explicar el retard.
flights = flights.drop("Diverted", axis=1) #No aporta informació rellevant per l'anàlisis.
```

#### 2.4 Fes una taula de les aerolínies amb més endarreriments acumulats.

```python
print('Aerolínies amb més endarreriments acumulats:\n')

delay_minutes_per_carrier = flights.groupby('UniqueCarrier')['ArrDelay'].sum()
delay_minutes_per_carrier.sort_values(ascending = False).head().to_frame()
```

#### 2.5 Quins són els vols més llargs?

```python
print("Informació dels vols més llargs:\n")

longest_flights = flights.sort_values(by='Distance', ascending=False)
longest_flights[['Distance', 'UniqueCarrier', 'FlightNum', 'TailNum', 'Origin', 'Dest']].head()
```

Els vols més llargs són els que fa l'aerolinia CO de l'aeroport Internacional de Newark Liberty (Nova Jersey) a l'aeroport  Internacional Daniel K. Inouye (Honolulu).

#### I els més endarrerits? 

```python
print("Informació dels vols més endarrerits:\n")

highest_delays = flights.sort_values(by='ArrDelay', ascending=False)
highest_delays[['ArrDelay', 'UniqueCarrier', 'FlightNum', 'TailNum', 'Origin', 'Dest']].head()
```

Aquí veiem els vols més endarrerits. El major retard és de 1382 minuts.

#### Busca les rutes més llargues 

```python
longest_routes = longest_flights[['Origin', 'Dest', 'Distance']]

# Ordenem les columnes 'Origin' i 'Dest' per evitar que les rutes en direcció inversa es considerin com a dues rutes diferents.
longest_routes[['Origin', 'Dest']] = longest_routes.apply(lambda row: sorted([row['Origin'], row['Dest']]), axis=1, result_type='expand')

longest_routes = longest_routes.drop_duplicates()

print('Rutes més llargues:\n')
longest_routes.head()
```

Tal i com podem veure les dues rutes més llargues tenen com a punt d'origen o destí Honolulu.

#### Busca les rutes que acumulen més retards¶

```python
delays_per_route = highest_delays.groupby(['Origin', 'Dest'])['ArrDelay'].sum().reset_index()

highest_delays[['Origin', 'Dest']] = highest_delays.apply(lambda row: sorted([row['Origin'], row['Dest']]), axis=1, result_type='expand')

delays_per_route = delays_per_route.drop_duplicates().sort_values(by='ArrDelay', ascending=False)

print('Retards acumulats per ruta:\n')
delays_per_route.head()
```

La ruta de l'aeroport LaGuardia (Nova York) a l'aeroport Internacional O'Hare (Chicago), en els dos sentits, és la que acumula més retards.

#### 2.6 Aporta allò que consideris rellevant.

```python
print('Retards acumulats segons el dia de la setmana:\n')
      
delays_per_dow = flights.groupby('DayOfWeek')['ArrDelay'].sum().to_frame()
delays_per_dow.sort_values(by='ArrDelay', ascending=False)
```

El dia amb més retards acumulats és el divendres, seguit del diumenge, coincidint amb els dies en els que la gent sol viatjar durant el cap de setmana. El dia amb menys retards és el dissabte.

```python
print('Contribució de cada causa al retard:\n')

flights_with_delays = flights[flights['Delay'] == 1]
delay_data = flights_with_delays[['ArrDelay','CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']]
delay_proportions = delay_data.div(delay_data['ArrDelay'], axis=0) * 100
delay_proportions.mean().to_frame().rename(columns={0: 'Delay_contribution'})
```

Les causes que més contribueixen al retard en els vols són: la propagació del retard (degut al retard en el vol anterior), amb casi un 40%, i el retard associat a l'aerolínia, amb un 32,7%. La seguretat és, amb diferència, la causa que menys contribueix al retard total, ramb menys d'un 1%.

```python
print('Retards per aerolínia:\n')

delays_per_carrier = flights.groupby('UniqueCarrier')['Delay'].sum()
delays_per_carrier.sort_values(ascending = False).head().to_frame().rename(columns={'Delay': 'Total_Delays'})
```

Les 5 aerolínies amb més retards coincideixen amb les aerolinies amb un major retard acumulat (en minuts). La única diferència està en las aerolínies United Airlines (UA) i American Eagle Airlines (MQ), doncs l'aerolínia UA és la tercera amb més minuts de retard acumulats però la quarta amb més retards. En el cas de l'aerolínia MQ succeix el contrari.

Per tant, encara MQ té més retards que UA, la segona acumula més minuts de retard.
